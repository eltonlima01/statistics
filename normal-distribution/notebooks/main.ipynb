01. Uma empresa produz televisores de dois tipos, tipo A (comum) e tipo B (luxo), e garante a restituição da quantia paga se qualquer televisor apresentar defeito grave no prazo de seis meses. O tempo para ocorrência de algum defeito grave nos televisores tem distribuição normal sendo que, no tipo A, com média de 10 meses e desvio padrão de 2 meses e no tipo B, com média de 11 meses e desvio padrão de 3 meses. Os televisores de tipo A e B são produzidos com lucro de 1200 u.m. e 2100 u.m. respectivamente e, caso haja restituição, com prejuízo de 2500 u.m. e 7000 u.m., respectivamente.

(a) Calcule as probabilidades de haver restituição nos televisores do tipo A e do tipo B.

<hr>

Primeiramente organizamos as informações dos tipos de televisores:

$
\text{tipo A (comum)} =
\begin{cases}
\mu    \text{ (média)}         & = 10 \\
\sigma \text{ (desvio padrão)} & = 2
\end{cases}
$

$
\text{tipo B (luxo)} =
\begin{cases}
\mu    & = 11 \\
\sigma & = 3
\end{cases}
$

A partir da média e do desvio padrão, podemos definir o Z-Score de cada tipo de televisor.\
Sendo a equação do Z-Score:

$ \boxed{Z = \frac{x - \mu}{\sigma}} $.

Temos, para o tipo A, o Z-Score definido como:

$ \boxed{Z_{A} = \frac{x - 10}{2}} $.

E para o tipo B:

$ \boxed{Z_{B} = \frac{x - 11}{3}} $.

Com o Z-Score dos tipos definidos, podemos agora calcular as probabilidades de haver restituição de cada tipo de televisor.\
Como a garantia de restituição ocorre caso algum televisor apresentar defeito no prazo de seis meses, podemos interpretar da seguinte maneira:

$ \text{Se algum defeito grave ocorrer em até 6 meses, ocorre restituição.} $

Queremos portanto a probabilidade de ocorrer algum defeito em até 6 meses (ou seja, de 6 meses pra baixo).\
Matematicamente, sendo $x$ a **probabilidade de ocorrer algum defeito**, escrevemos como:

$ \boxed{P (x \leq 6)} $

$ \text{x menor ou igual à 6 (meses)} $.

Substituimos então $x$ nas equações de Z-Score:

$ Z_{A} = \frac{6 - 10}{2} = -4 / 2 $

$ \boxed{Z_{A} = -2} $.

Consultando a **tabela da distribuição normal unicaudal**, temos para o Z-Score $2$ o valor $\boxed{0.4772}$.

$ Z_{B} = \frac{6 - 11}{3} = -5 / 3 $

$ \boxed{Z_{B} = -1.67} $.

Consultando a tabela, temos para o Z-Score $1.67$ o valor $\boxed{0,4525}$.

A partir dessas informações, partimos agora para a análise do gráfico da distribuição normal.

Começando para os **televisores do tipo A**:


```python
import matplotlib.pyplot as plt
import numpy as np

x1, x2, points = 0, 20, 1000
x = np.linspace (x1, x2, points)

mu = 10
sigma = 2

y = np.exp (- ((x - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

i1, i2 = 6, 10
i = np.linspace (i1, i2, points)

yi = np.exp (- ((i - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.figure (figsize = [10, 10])

plt.subplot (2, 1, 1)
plt.plot (x, y)
plt.xlabel ("$x$")
plt.title (r"$\text{Distribuição normal para } \mu = 10 \text{ e } \sigma = 2$")
plt.grid ()

plt.subplot (2, 1, 2)
plt.plot (x, y)
plt.xlabel ("$x$")
plt.title (r"$\text{Área para x entre 6 e 10}$")
plt.fill_between (i, yi)
plt.grid ()

plt.show ()
```

O exercício pede a probabilidade de ocorrer restituição. A restituição ocorre caso aconteça algum defeito grave nos televisores num prazo de **até 6 meses**, ou seja, no gráfico, estamos falando de valores de valores de $x$ **de $6$ pra baixo**:

```python
x1, x2 = 0, 10
x = np.linspace (x1, x2, points)

y = np.exp (- ((x - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.plot (x, y)
plt.grid ()
plt.xlabel ("$x$")

i1, i2 = 0, 6
i = np.linspace (i1, i2, points)

yi = np.exp (- ((i - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.fill_between (i, yi, color = "red")

plt.show ()
```

Portanto a situação atual é:

Com o Z-Score calculamos a área do gráfico para os valores de $x$ **entre 6 e 10**, mas para encontrarmos a probabilidade de ocorrer restituição em até 6 meses, precisamos da **área do gráfico para valores de $x$ menores ou iguais a 6**.

```python
x1, x2 = -2.5, 12.5
x = np.linspace (x1, x2, points)

y = np.exp (- ((x - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.plot (x, y)
plt.grid ()
plt.xlabel ("$x$")

i1, i2 = 10, 6
i = np.linspace (i1, i2, points)

yi = np.exp (- ((i - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.fill_between (i, yi, label = "Área para x entre 6 e 10")

i1, i2 = 6, -2.5
i = np.linspace (i1, i2, points)

yi = np.exp (- ((i - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.fill_between (i, yi, color = "red", label = r"Área para x menor ou igual a 6")

plt.legend (loc = "upper left")
plt.show ()
```

Para encontrarmos a área vermelha (valores de x menores ou iguais a 6), precisaremos utilizar de algumas propriedades do gráfico da distribuição normal.

**A área total sob a curva vale 1**.

Ou seja, toda a área do gráfico da distribuição normal vale 1 (se $x$ são as possibilidades, para todos os $x$ teríamos $100\%$ de possibilidades).

```python
x1, x2 = 0, 20
x = np.linspace (x1, x2, points)

y = np.exp (- ((x - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.figure (figsize = [10, 5])

plt.plot (x, y)
plt.grid ()
plt.xlabel ("$x$")

i1, i2 = 0, 20
i = np.linspace (i1, i2, points)

yi = np.exp (- ((i - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.fill_between (i, yi, color = "green", label = "Área sob toda a curva = 1 (100%)")

plt.legend (loc = "upper right")
plt.show ()
```

**A curva é simétrica em torno de $x = média$**.

Como a área sob toda a curva vale 1 (100%), se dividirmos a área no centro exato do gráfico, teremos exatamente 0.5 (50%) da área no lado esquerdo, e 0.5 (50%) da área no lado direito.

```python
x1, x2 = 0, 20
x = np.linspace (x1, x2, points)

y = np.exp (- ((x - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.figure (figsize = [11, 6])

plt.plot (x, y)
plt.grid ()
plt.xlabel ("$x$")

i1, i2 = 0, 10
i = np.linspace (i1, i2, points)

yi = np.exp (- ((i - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.fill_between (i, yi, color = "green", label = "Área sob o lado esquerdo da curva = 0.5 (50%)")

plt.legend (loc = "upper right")
plt.show ()
```

Voltando para a questão inicial, temos a área para x entre 6 e 10 ($0.4772 = 47.72\%$).

Se a área do lado esquerdo do gráfico é 0.5 ($50\%$), temos portanto:

$ \text{Área que não temos (para x menor ou igual a 6)} + \text{Área que temos (para x entre 6 e 10 = 0.4772)} = 0.5 (50\% \text{ da área}) $

Logo

$ \text{Área que não temos} + 0.4772 = 0.5 $

$ \text{Área que não temos} = 0.5 - 0.4772 $

$ \text{Área que não temos} = 0,0228 $

Ou seja, a área para valores de $x$ menores ou iguais $6$ é $0.0228$ ($2.28\%$).

```python
x1, x2 = -2.5, 12.5
x = np.linspace (x1, x2, points)

y = np.exp (- ((x - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.figure (figsize = [10, 10])
plt.plot (x, y)
plt.grid ()
plt.xlabel ("$x$")

i1, i2 = 10, 6
i = np.linspace (i1, i2, points)

yi = np.exp (- ((i - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.fill_between (i, yi, label = "Área que temos (para x entre 6 e 10) = 0.4772 (47.72%)")

i1, i2 = 6, -2.5
i = np.linspace (i1, i2, points)

yi = np.exp (- ((i - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.fill_between (i, yi, color = "red", label = "Área que acabamos de calcular (para valores de x menores ou iguais 6) = 0.0228 (2.28%)")

plt.legend (loc = "upper left")
plt.show ()
```

**Recapitulando:**

O tempo para ocorrência de algum defeito grave nos televisores do tipo A (normal) tem média de 10 meses e desvio padrão de 2 meses.\
Portanto o Z-Score para o tipo A:

$ \boxed{Z_{A} = \frac{x - 10}{2}} $

Se ocorrer algum defeito em até 6 meses, ocorre também restituição da quantia paga.\
O exercício pede para calcularmos a probabilidade de restituição de cada tipo de televisor, queremos portanto a probabilidade de ocorrer algum defeito grave em até 6 meses:

$ \boxed{P (x \leq 6)}$.

$ Z_{A} = \frac{6 - 10}{2} $

$ Z_{A} = -4 / 2 $

$ \boxed{Z_{A} = -2} $

Na tabela, um Z-Score de $2$ tem o valor $0.4772$, portanto essa é a área que temos entre 6 e 10 ($47.72\%$).

Se a área do lado esquerdo do gráfico equivale a $0.5 (50\%)$ da área total, temos essa equação:

$ \text{Área que não temos (para x menor ou igual a 6)} + \text{Área que temos (para x entre 6 e 10 = 0.4772)} = 0.5 (50\%) $

$ \text{Área que não temos} + 0.4772 = 0.5 $

$ \text{Área que não temos} = 0.5 - 0.4772 $

$ \text{Área que não temos} = 0.0228 $

Portanto, a probabilidade de ocorrer algum defeito grave em até 6 meses é:

$ \boxed{P(x \leq 6) = 0.0228 (2.28\%)} $.

$ \boxed{\text{A probabilidade de ocorrer algum defeito grave em até 6 meses para os televisores do tipo A (normal), e consequentemente ocorrer restituição, é de } 2.28\%} $.

Para os televisores do tipo B (luxo), simplesmente **repetimos o exato mesmo processo**.

$ \boxed{P (x \leq 6)}$.

$ Z_{B} = \frac{6 - 11}{3} $

$ Z_{B} = -5 / 3 $

$ \boxed{Z_{B} = -1.67} $

Na tabela, um Z-Score de $1.67$ tem o valor $0,4525$, portanto essa é a área que temos entre 6 e 11 ($45.25\%$).

Se a área do lado esquerdo do gráfico equivale a $0.5 (50\%)$ da área total, temos essa equação:

$ \text{Área que não temos (para x menor ou igual a 6)} + \text{Área que temos (para x entre 6 e 10 = 0.4525)} = 0.5 (50\%) $

$ \text{Área que não temos} + 0.4525 = 0.5 $

$ \text{Área que não temos} = 0.5 - 0.4525 $

$ \text{Área que não temos} = 0.0475 $

Portanto, a probabilidade de ocorrer algum defeito grave em até 6 meses é:

$ \boxed{P(x \leq 6) = 0.0475 (4.75\%)} $.

$ \boxed{\text{A probabilidade de ocorrer algum defeito grave em até 6 meses para os televisores do tipo B (luxo), e consequentemente ocorrer restituição, é de } 4.75\%} $.

2. A concentração de um poluente em água liberada por uma fábrica tem distribuição N(8; 2,25). Qual a chance, de que num dado dia, a concentração do poluente exceda o limite regulatório de 10 ppm?

<hr>

A notação $X \sim N(\mu, \sigma^{2})$ significa que a $\text{a variável aleatória X segue segue uma distribuição normal com média } \mu \text{ e variância } \sigma^{2}$.

Portanto, para $X \sim N(8, 2.25)$, temos:

$
\begin{cases}
\mu        \text{ (média)}     & = 8 \\
\sigma^{2} \text{ (variância)} & = 2.25
\end{cases}
$

Note que na equação do Z-Score **não utilizamos a variância ($\sigma^{2}$), mas sim o desvio padrão ($\sigma$)**:

$ \boxed{Z = \frac{x - \mu}{\sigma}} $.

Transformar a variância no desvio padrão se reduz simplesmente à **calcular a raíz quadrada da variância ($\sigma^{2}$)**:

$ \sigma^{2} = 2.25 $

$ \sigma = \sqrt{2.25} $

$ \sigma = 1.5 $

Obtemos então os parâmetros ($\mu$, $\sigma$) para a equação do Z-Score:

$ \boxed{Z = \frac{x - 8}{1.5}} $.

Podemos então partir agora para a resolução do exercício, que pede a **probabilidade da concentração do poluente exceda o limite de 10 ppm**.

Com "exceder o limite de 10 ppm", queremos a probabilidade para valores de x **maiores que 10**:

$ \boxed{P(x > 10)} $.

No gráfico:

```python
x1, x2 = 0, 16
x = np.linspace (x1, x2, points)

mu = 8
sigma = 1.5

y = np.exp (- ((x - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

i1, i2 = 10, 16
i = np.linspace (i1, i2, points)

yi = np.exp (- ((i - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.figure (figsize = [10, 10])

plt.subplot (2, 1, 1)
plt.plot (x, y)
plt.xlabel ("$x$")
plt.title (r"$\text{Distribuição normal para } \mu = 8 \text{ e } \sigma = 1.5$")
plt.grid ()

plt.subplot (2, 1, 2)
plt.plot (x, y)
plt.fill_between (i, yi, color = "red")
plt.xlabel ("$x$")
plt.title (r"$\text{Área para x maior que 10}$")
plt.grid ()

plt.show ()
```

Primeiraente aplicamos $x = 10$ na equação do Z-Score:

$ Z = \frac{10 - 8}{1.5} = 2 / 1.5 $

$ Z = 1.33 $

Consultando a **tabela da distribuição normal unicaudal**, temos para um Z-Score de $1.33$, o valor $\boxed{0,4082}$.\
Esse valor representa a quantidade de área que existe entre a parcela do gráfico para valores de $x$ 8 e 10 ($40.82\%$).

```python
plt.plot (x, y)
plt.xlabel ("$x$")
plt.grid ()

i1, i2 = 8, 16
i = np.linspace (i1, i2, points)

yi = np.exp (- ((i - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.fill_between (i, yi, label = r"$\text{Área para x entre 8 e 10}$")

i1, i2 = 10, 16
i = np.linspace (i1, i2, points)

yi = np.exp (- ((i - mu) ** 2) / ((2 * sigma ) ** 2)) / sigma * np.sqrt (2 * np.pi)

plt.fill_between (i, yi, color = "red", label = r"$\text{Área para x maior que 10}$")

plt.legend (loc = "upper left")
plt.show ()
```

Se a área do lado direito do gráfico é 0.5 ($50\%$), temos portanto:

$ \text{Área que não temos (para x maior que 10)} + \text{Área que temos (para x entre 8 e 10 = 0,4082)} = 0.5 \text{ (50\% da área)} $.

Logo:

$ \text{Área que não temos} + 0.4082 = 0.5 $

$ \text{Área que não temos} = 0.5 - 0.4082 $

$ \text{Área que não temos} = 0.0918 $

Ou seja, a área para valores de $x$ maiores que $10$ é $0.0918$ ($9.18\%$).

Portanto,

$ \boxed{\text{a probabilidade de que num dado dia a concentração do poluente exceda o limite regulatório de 10 ppm é de } 9.18\%} $.
