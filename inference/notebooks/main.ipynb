## Atividade 1

- Uma empresa deseja estimar o tempo médio que seus funcionários gastam para concluir uma tarefa administrativa. Para isso, foram analisados os tempos de execução de apenas 40 funcionários escolhidos aleatoriamente.\
Qual alternativa representa corretamente o conceito de inferência estatística nesse contexto?
    - Calcular exatamente o tempo de todos os funcionários da empresa.
    - Utilizar os dados da amostra para fazer conclusões sobre toda a população permanentemente.
    - Eliminar a variabilidade presente nos dados coletados.

<hr>

**Utilizar os dados da amostra para fazer conclusões sobre toda a população permanentemente.**

Nesse contexto, fazer conclusões sobre todos os funcionários da empresa a partir dos resultados observados dos 40 funcionários escolhidos aleatoriamente melhor representa o conceito de inferência estatística.

> Definição
> 
> Processo de obter informações sobre uma população a partir de resultados observados na amostra. A essência: generalizar resultados amostrais para toda a população.

- Uma empresa de segurança digital monitora tentativas de login em um servidor. Em 12 acessos independentes, observou-se sucesso em 9 deles.\
Assumindo um modelo binomial, o estimador de máxima verossimilhança para a probabilidade de sucesso é:
    - 0,25
    - 0,5
    - 0,75
    - 0,9

<hr>

**0,75**

A função de verossimilhança é:

$ L(p) = P(9 \text{ sucessos e } 3 \text{ fracassos}) = p^{9}(1 - p)^3 $.

Maximizando essa função em relação a $ p $ (derivando), obtemos:

$ 9 p^{8} (1 - p)^{3} - 3 p^{9} (1 - p)^{2} $

Igualando a zero:

$ 9 p^{8} (1 - p)^{3} - 3 p^{9} (1 - p)^{2} = 0 $

$ 3*3 p^{8} (1 - p)^{3} - 3 p^{9} (1 - p)^{2} = 0 $

$ 3 [3 p^{8} (1 - p)^{3} - p^{9} (1 - p)^{2}] = 0 $

$ 3 p^{8} (1 - p)^{3} - p^{9} (1 - p)^{2} = 0 $

$ 3 p^{8} (1 - p)^{3} - p^{8} * p (1 - p)^{2} = 0 $

$ p^{8} [3 (1 - p)^{3} - p (1 - p)^{2}] = 0 $

$ 3 (1 - p)^{3} - p (1 - p)^{2} = 0 $

$ 3 (1 - p)^{2} * (1 - p) - p (1 - p)^{2} = 0 $

$ (1 - p)^{2} [ 3 (1 - p) - p ] = 0 $

$ 3 (1 - p) - p = 0 $

$ 3 - 3p - p = 0 $

$ 3 - 4p = 0 $

$ 3 = 4p $

$ \boxed{ p = 3/4 = 0,75 } $

De modo geral, o EMV do parâmetro $ p $ de uma distribuição binomial é:

$ \hat{p}_{emv} = \frac{x}{n} $.

Logo:

$ \hat{p}_{emv} = 9/12 = 0,75 $.

- Um pesquisador deseja estimar o peso médio de drones produzidos por uma empresa. Foi retirada uma amostra aleatória de 16 drones, obtendo média amostral de 2,4 kg e desvio-padrão amostral de 0,3 kg. Assume-se normalidade dos dados.\
Qual procedimento é mais adequado para construir o intervalo de confiança da média?
    - Utilizar a distribuição normal padrão ($ Z $), pois a população é normal.
    - Utilizar a distribuição t-Student, pois o desvio-padrão populacional é desconhecido e $ n < 30 $.
    - Utilizar a distribuição qui-quadrado, pois o parâmetro estimado é a média.

<hr>

**Utilizar a distribuição t-Student, pois o desvio-padrão populacional é desconhecido e $ n < 30 $.**

Como a amostra é pequena, o desvio-padrão populacional é desconhecido e a população original tem distribuição essencialmente normal, o procedimento mais adequado para construir o intervalo de confiança da média é a distribuição t-Student.

> **Quando Usar a Distribuição t?**
>
> **Amostra Pequena**
>> Tamanho da amostra $ n \leq 30 $
>
> **$ \sigma $ Desconhecido**
>> Desvio-padrão populacional não disponível
>
> **Normalidade da População**
>> A população original tem distribuição essencialmente normal (verificar com histograma ou teste de Shapiro-Wilk)

- Uma pesquisadora está analisando dois estimadores g1 e g2 para estimar a média populacional do consumo de energia de um equipamento.\
Após vários estudos, observou-se que: Ambos os estimadores são não tendenciosos; O estimador g1 possui variância igual a 2; O estimador g2 possui variância igual a 5.\
Com base nas propriedades dos estimadores, assinale a alternativa correta:
    - O estimador g2 é mais eficiente, pois possui maior variância.
    - O estimador g1 é mais eficiente, pois apresenta menor variância entre estimadores não viesados.
    - Ambos possuem a mesma eficiência, pois são não tendenciosos.
    - Nenhum dos estimadores pode ser considerado adequado.

<hr>

**O estimador g1 é mais eficiente, pois apresenta menor variância entre estimadores não viesados.**

Propriedade:

> **Eficiência**
>> Refere-se à variância dos estimadores. Se g é não viesado (ou não tendencioso) e tem a menor variância na classe dos estimadores não viesados, diz-se que g é um estimador eficiente.

Como ambos g1 e g2 são não viesados, o estimador mais eficiente é o que apresenta menor variância (g1).

## Atividade 2

- Considere as afirmações abaixo sobre erros em testes de hipóteses:\
I. Erro Tipo I ocorre quando rejeitamos $ H_{0} $ sendo ela verdadeira.\
II. Erro Tipo II ocorre quando não rejeitamos $ H_{0} $ sendo ela falsa.\
III. O nível de significância α representa a probabilidade de cometer erro Tipo I.\
Assinale a alternativa correta:

    - Apenas I é verdadeira.
    - Apenas II é verdadeira.
    - Apenas I e II são verdadeiras.
    - I, II e III são verdadeiras.

<hr>

**I, II e III são verdadeiras.**

|                                           | $ H_{0} $ é verdadeira                                          | $ H_{0} $ é falsa                                               |
| :---:                                     | :---:                                                           | :---:                                                           |
| Rejeitar $ H_{0} $ (Significante)         | Erro Tipo I (Rejeitar $ H_{0} $, quando $ H_{0} $ é verdadeira) | Decisão Correta                                                 |
| Não Rejeitar $ H_{0} $ (Não Significante) | Decisão Correta                                                 | Erro Tipo II (Não Rejeitar $ H_{0} $, quando $ H_{0} $ é falsa) |

Portanto, **I e II são verdadeiras**.

$ \alpha = P(\text{Erro Tipo I}) = P(\text{Rejeitar } H_{0} | H_{0} \text{ é verdadeira}) $

$ \alpha - \text{ Nível de significância} $

Portanto I, II e III são verdadeiras.

- Uma indústria afirma que o peso médio de seus pacotes é de 500 g. Sabe-se que o desvio-padrão populacional é de 20 g.\
Uma amostra aleatória de 64 pacotes apresentou média de 495 g.\
Considere o teste:\
$ H_{0} $: $ \mu $ = 500 $ H_{1} $: $ \mu $ < 500 $ \alpha $ = 0,05\
Qual é o valor da estatística de teste Z?

    - -1,00
    - -2,00
    - -2,50
    - -4,00

<hr>

**-2,00**

$ \boxed{Z_{0} = \frac{\bar{x} - \mu_{0}}{\sigma / \sqrt{n}}} $

$ Z_{0} = \frac{495 - 500}{20 / \sqrt{64}} = \frac{-5}{20/8} = - 5/2,5 $

$ \boxed{Z_{0} = -2} $

- Os registros de uma escola indicam média histórica de 70 pontos em uma prova.\
Uma amostra de 25 alunos apresentou:\
média = 75 pontos desvio-padrão = 10 pontos\
Deseja-se testar:\
$ H_{0}: \mu = 70 $ $ H_{1}: \mu \neq 70 $\
Qual o valor da estatística t calculada?

    - 1,50
    - 2,00
    - 2,50
    - 3,00

<hr>

**2,50**

$ \boxed{T = \frac{\bar{X} - \mu}{S / \sqrt{n}}} $

$ T = \frac{75 - 70}{10 / \sqrt{25}} = \frac{5}{10/5} = 5 / 2 $

$ \boxed{T = 2,5} $
